## Task 1: Our first task is to first identify the three best-performing industries based on the number of new unicorns created over the last three years (2019, 2020, and 2021) combined. Let's begin by writing a query that: 

- Fetches the industry 
- The count of unicorns added in that industry in the years 2021, 2020 & 2019
- And, let's order the result in descending order. 

```python
SELECT 	industry,
		COUNT(*)
FROM industries
INNER JOIN dates AS d
	USING(company_id)
WHERE EXTRACT(year FROM d.date_joined) IN (2019, 2020, 2021)
GROUP BY 1
ORDER BY COUNT(*) DESC
LIMIT 3;
```

So, now we know that the top three best-performing industires are:
- Fintech
- Internet software & services
- E-commerce & direct-to-consumer

We will be using this result in our next query. 

## Task 2: Our second task is to write a query to return the industry, the year, the number of companies in these industries that became unicorns each year in 2019, 2020, and 2021, along with the average valuation per industry per year, converted to billions of dollars and rounded to two decimal places.

- First, let's write a Common Table Expression called unicorns to retrieve the industry, year and number ofd unicorns in that industry. We need to filter this in order to show the data pertaining to the top three industries we calculated above.

- Then we will join this table with the funding table to calculate the final result. 

```python
WITH unicorns AS (
SELECT 	industry,
		company_id,
    	EXTRACT(year FROM d.date_joined) AS year_f
FROM industries
INNER JOIN dates AS d
	USING(company_id)
WHERE EXTRACT(year FROM d.date_joined) IN (2019, 2020, 2021)
AND industry IN ('Fintech', 'Internet software & services', 'E-commerce & direct-to-consumer')
)

SELECT 	industry,
		year_f AS year,
        COUNT(company_id) AS num_unicorns,
        ROUND(AVG(f.valuation/1000000000), 2) AS average_valuation_billions
FROM unicorns
INNER JOIN funding AS f
	USING(company_id)
GROUP BY industry, year_f
ORDER BY industry, year_f DESC;        
```

## Summary of Results

- Fintech, Internet software & services & E-commerce & direct-to-consumer are the three top performing industries as they have the most number of unicorns in the past 3 years. 

- All three industries have shown a significant increase in the number of new unicorns added in 2021.

- In 2021, the Fintech industry had the most number of new unicorns added with an avg valuation of $2.75B.
